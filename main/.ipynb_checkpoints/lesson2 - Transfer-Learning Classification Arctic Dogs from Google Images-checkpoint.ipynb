## Let's View our Data!
My preference is to have the same random images everytime we run, therefore we can use numpy seed to do this

````python
np.random.seed(42) # Seed our arbitrary chosen images

# Use ImageDataBunch for training purposes
# 1. We will use .from_folder method since we put all of our images in a folder
# 2. We specify the path of our main directory 'data/arctic_dogs'
# 3. We specify the percentage of data that we want to use for validation data in this case 20%
# 4. We will use get_transforms() to transform the images to have the same format. This will be a tuple of train and validation
# 5. We specify the size of the image in this case 224x224 pixels
# 6. We specify the number of process to launch for data collection in this case we use 4 processes.
# 7. Lastly we will normalize the data into a a tensor.
data = ImageDataBunch.from_folder(
                      path, 
                      train=".",
                      valid_pct=0.2,
                      ds_tfms=get_transforms(),
                      size=224,
                      num_workers=4).normalize(imagenet_stats)
````

````python
# Lets take a look at the classes
data.classes
````

````python
data.show_batch(rows=3, figsize=(10,12))
````

````python
# Show the data classes
print(data.classes)
# Show the number of labels that we have
print(data.c)
# Show the length of the training data set
print(len(data.train_ds))
# Show the length of the validation data set
print(len(data.valid_ds))
````

## Let's Train our Model!

````python
# Let's use Resnet34 Architecture for our first training, and show the error rate for every training cycles.
learn = create_cnn(data, models.resnet34, metrics=error_rate)
````

````python
# Let's do 4 cycles and see how good is our model
learn.fit_one_cycle(4)
````

As we can see, the error rate is pretty high finishing at 18% error rate and the end of cycle 4 training cycle. This model will need to be re-tune in order to minimize our error to achieve state of the art classifier.

````python
learn.save('stage-1')
````

Unfreeze the model and find the optimal learning rate for next training. Remember learning rate will help the model to optimize better to find the global min of the error function. We can use the lr_find() function in our learning model and plot the graph

````python
learn.lr_find()
````

````python
# Let's plot the graph
learn.recorder.plot()
````

### By convention we need to consider these factors when we are trying to find the optimal learning rate
1. Find the learning rate (x-axis) before the Loss value (y-axis) flies of the roof. (1e-04)
2. Then find the learning rate 10x smaller. (1e-05)
3. We will use slice method (1e-05) for our learning rate

````python
# Train the model with our new learning rate
learn.fit_one_cycle(2, max_lr=slice(1e-05))
````

We can see that the model is training better after 2 cycles, we can train it again to see if there's any difference

````python
learn.fit_one_cycle(5, max_lr=slice(1e-05))
````

Save the model and freeze it for future tuning
## Model behavior Interpretation
We can use the ClassificationInterpretation class from FastAI to see which classes have the most errors and missclasification so that we can fine tune our data, learning rate, training cycles and our data transformation itself.

````python
learn.save('stage-2')
````

````python
learn.load('stage-2')
````

````python
interpretation = ClassificationInterpretation.from_learner(learn)
````

````python
# Plot the confusion matrix to see where does the most errors are made
interpretation.plot_confusion_matrix()
````

We can clearly see that most of the errors was between alaskan malamute and siberian husky, our model train really well on predicting Samoyed and Akita.
## Cleaning Up Data
What we can do here to optimize our model is to remove images that is irrelevant in our dataset. We can manually delete these files in the directory

````python
# Get the top losses, that has the worse error and the indexes of these images
losses, indexes = interpretation.top_losses()
# Get the paths of these highes losses images from our validation data set
top_loss_paths = data.valid_ds.x[indexes]
# Print the paths of these images
print(top_loss_paths)
````

Delete the files shown above to optimize our model

## Let's Do Transfer Learning to optimize our Model
One of the tricks to optimize our model is to start with smaller image sizes and transfer the learned weights to a new data set of bigger image sizes and see the difference in error rate while optimizing in learning rates as well.

````python
# Create a new data set with 128 as the size of the images
data_smaller = ImageDataBunch.from_folder(path,
                                          train=".",
                                          valid_pct=.2,
                                          ds_tfms=get_transforms(),
                                          size=128,
                                          num_workers=4).normalize(imagenet_stats)
````

````python
data_smaller.show_batch(rows=3, figsize=(7,8))
````

````python
# Create a new cnn to train with this data set
learn = create_cnn(data_smaller, models.resnet34, metrics=error_rate)
````

````python
# Train for 4 cycles
learn.fit_one_cycle(4)
````

We can really see the difference here, previously we have an error rate of roughly 19% now after we change the size of the image we can confidently say that the model is training better.

````python
learn.save('stage-1-128')
````

````python
# Let's find the best learning rate
learn.unfreeze()
learn.lr_find()
````

````python
# Plot the learning rate
learn.recorder.plot()
````

````python
# Train the model again using the same convention as before
lr = 1e-04
learn.fit_one_cycle(2, slice(lr))
````

Wow! After finding using the new learning rate we can see the difference in the error rate by 2% decrease. Now let's begin the transfer learning by creating a new data with the size of 256 to see the difference.

````python
data_bigger = ImageDataBunch.from_folder(path,
                                        train=".",
                                        valid_pct=.2,
                                        ds_tfms=get_transforms(),
                                        size=256,
                                        num_workers=4).normalize(imagenet_stats)
````

````python
# Update the learn data to use this bigger size data
learn.data = data_bigger
````

````python
# Unfreeze() the model and look for learning rates and train!
learn.unfreeze()
lr_find(learn)
learn.recorder.plot()
````

### Tweaking Learning Rates
By convention we can slice the learning rate by using the newly founded learning rate with the previously found learning rate when the data was in size 128. slice(new learning rate, previous learning rate / 5), we divide the previous learning rate by 5 to find the middle point of the maximum slice

````python
previous_lr = 1e-05
learn.fit_one_cycle(3, slice(1e-04, previous_lr/5))
````

````python
learn.save('stage-2-transfer')
````

## Conclusion
We can see the significance difference here. Way back from using 224 image size we have a best error rate of roughly 19%. After conducting a transfer learning implementation, we lower the error rate from roughly 14% to 9% error rate, giving us roughly 91% accuracy of classifiying arctic dogs! 

We can see the confusion matrix and see the difference in error in comparison to our first model

````python
learn.load('stage-2-transfer')
interpretation = ClassificationInterpretation.from_learner(learn)
interpretation.plot_confusion_matrix()
````

We can see that the difference is huge in compare to our first model, this time we only made 3 mistakes in classifiying a siberian husky with alaskan malamute. Previously the model mistakenly predicted 17 times and it got decreased significantly.
